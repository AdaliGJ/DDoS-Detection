```python
import pandas as pd
chunk_size = 100_000  # Adjust chunk size
DDoS_DNS_path = "datos/CSV-01-12/01-12/DrDos_DNS.csv"  # File path

# Create an empty DataFrame to store the data
DDoS_DNS = pd.DataFrame()
BENIGN = pd.DataFrame()

# Iterate over the file in chunks
for chunk in pd.read_csv(DDoS_DNS_path, chunksize=chunk_size):
    # Process the chunk of data
   
    processed_chunk = chunk[chunk[' Label'] != 'BENIGN'] 

    benign_chunk = chunk[chunk[' Label'] == 'BENIGN']
    
    # Append the processed chunk to the DataFrames
    DDoS_DNS = pd.concat([DDoS_DNS,processed_chunk])
    BENIGN = pd.concat([BENIGN,benign_chunk])

    #Free up memory by deleting the chunks if they're no longer needed
    del chunk
    #break


print("DataFrame shape:", BENIGN.shape)


BENIGN.head()
```

```python
sampled_rows = DDoS_DNS.sample(n=56863, random_state=42)

sampled_rows.to_csv('datos/standarizedData/DNS_DDoS.csv', index=False)
```

```python
import pandas as pd
import glob

# oBTENER TODOS LOS ARCHIVOS CSV CON LOS REGISTROS DE TRÁFICO
file_list = glob.glob('datos/standarizedData/*.csv')

# CREAR UNA LISTA PARA GUARDAR TODOS LOS DATAFRAMES
dfs = []

# LEER LOS ARCHIVOS CSV Y AÑADIRLOS A LA LISTA
for file in file_list:
    df = pd.read_csv(file)
    dfs.append(df)

# CONCATENAR LOS DATAFRAMES EN UNO SOLO
concatenated_df = pd.concat(dfs)

# GUARDAR EL DATASET BALANCEADO
concatenated_df.to_csv('datos/standarizedDataset.csv', index=False)
```
